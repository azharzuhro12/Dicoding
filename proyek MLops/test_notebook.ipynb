```python
import tensorflow as tf
import json
import base64
import requests


def prepare_example(tweet):
    # Define the features of your data
    feature = {
        "tweet": tf.train.Feature(bytes_list=tf.train.BytesList(value=[tweet.encode()])),
    }

    # Create an Example
    example = tf.train.Example(features=tf.train.Features(feature=feature))

    # Serialize the Example to a string
    example_str = example.SerializeToString()

    # Prepare the JSON data
    json_data = {
        "signature_name": "serving_default",
        "instances": [{"b64": base64.b64encode(example_str).decode()}]
    }

    return json.dumps(json_data)

tweet = "beautiful"

json_data = prepare_example(tweet)


endpoint = "http://103.190.215.87:8501/v1/models/cc-model:predict"
response = requests.post(endpoint, data=json_data)

prediction = response.json().get("predictions")
if prediction:
    prediction_value = prediction[0][0]
    result = "positif" if prediction_value > 0.6 else "negatif"
else:
    result = "Error: No predictions found in response."

print(result)
```
