```python
import joblib

# Memuat model terbaik
best_model = joblib.load('/content/Model_best_logistic.pkl')

# Memuat TfidfVectorizer yang sama
tfidf = joblib.load('/content/tfidf.pkl')

data_input = [
    "Saya sangat senang dengan produk ini!",
    "Saya kecewa sama barangya",
    "banyak barangya"
]

# Ekstraksi fitur dari data input
X_input_tfidf = tfidf.transform(data_input)

# Melakukan prediksi
predictions = best_model.predict(X_input_tfidf.toarray())

# Memperoleh probabilitas prediksi untuk setiap kelas
prediksi_probabilitas = best_model.predict_proba(X_input_tfidf.toarray())

# Label sentimen
label = ["NEGATIF", "NETRAL", "POSITIF"]
threshold_netral = 0.5

# Tampilkan hasil untuk setiap kalimat
for i, kalimat in enumerate(data_input):
    print(f"\nKalimat: {kalimat}")
    print("Probabilitas Sentimen:", prediksi_probabilitas[i])

    if prediksi_probabilitas[i][1] > threshold_netral:
        print("Sentimen: NETRAL")
    elif prediksi_probabilitas[i][2] > prediksi_probabilitas[i][0]:
        print("Sentimen: POSITIF")
    else:
        print("Sentimen: NEGATIF")
```
